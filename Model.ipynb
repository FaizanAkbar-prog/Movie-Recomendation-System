```python
import ast
def convert3(obj):
    count=0
    L=[]
    for i in ast.literal_eval(obj):
        if count!=3:
            L.append(i['name'])
            count=count+1
        else:
            break    
    return L    
```

```python
from nltk.stem.porter import PorterStemmer
ps=PorterStemmer()

def stem(text):
    y=[]
    for i in text.split():
        y.append(ps.stem(i))
    return ' '.join(y)
```

```python
new_movies['tags']=new_movies['tags'].apply(lambda x: x.lower())
```

```python
new_movies['tags']=new_movies['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer(stop_words='english',max_features=5000)

vectors = cv.fit_transform(new_movies['tags']).toarray()
```

```python
def recommend(movie):
    movie_index=new_movies[new_movies['title']== movie].index[0]
    distances=Cos[movie_index]
    movie_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]

    for i in movie_list:
        print(new_movies.iloc[i[0]].title)
```

```python
def precision_at_k(movie, k=5):
    # Get recommendations
    movie_index = new_movies[new_movies['title'] == movie].index[0]
    distances = Cos[movie_index]
    movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:k+1]
    recommendations = [new_movies.iloc[i[0]].title for i in movie_list]
    
    # Get input movie genres
    input_genres = movies[movies['title'] == movie]['genres'].values[0]
    input_genres = set(input_genres)
    
    relevant = 0
    for rec in recommendations:
        rec_genres = movies[movies['title'] == rec]['genres'].values
        if len(rec_genres) > 0:
            rec_genres = set(rec_genres[0])
            if len(input_genres.intersection(rec_genres)) > 0:
                relevant += 1
    
    precision = relevant / k
    return precision

print("Precision@5 for Avatar:", precision_at_k('Avatar'))
print("Precision@5 for Batman:", precision_at_k('Batman'))
print("Precision@5 for Inception:", precision_at_k('Inception'))
```

```python
def coverage(sample_size=100):
    import random
    all_movies = new_movies['title'].values
    sample_movies = random.sample(list(all_movies), sample_size)
    
    recommended_set = set()
    for movie in sample_movies:
        try:
            movie_index = new_movies[new_movies['title'] == movie].index[0]
            distances = Cos[movie_index]
            movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
            recs = [new_movies.iloc[i[0]].title for i in movie_list]
            for r in recs:
                recommended_set.add(r)
        except:
            pass
    
    coverage_score = len(recommended_set) / len(all_movies) * 100
    return coverage_score

print(f"Coverage: {coverage():.2f}%")
```

```python
test_movies = ['Avatar', 'Batman', 'Inception', 'Titanic', 'Interstellar']

print("="*45)
print("       MODEL EVALUATION REPORT")
print("="*45)

precisions = []

for movie in test_movies:
    try:
        p = precision_at_k(movie)
        precisions.append(p)
        
        movie_index = new_movies[new_movies['title'] == movie].index[0]
        distances = Cos[movie_index]
        movie_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x: x[1])[1:6]
        recs = [new_movies.iloc[i[0]].title for i in movie_list]
        
        print(f"\nMovie           : {movie}")
        print(f"Precision@5     : {p:.2f}")
        print(f"Recommendations : {recs}")
    except:
        print(f"\nMovie : {movie} not found, skipping...")

print("\n" + "="*45)
print(f"  Average Precision@5 : {sum(precisions)/len(precisions):.2f}")
print(f"  Coverage            : {coverage():.2f}%")
print("="*45)
```
